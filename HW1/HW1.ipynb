```python
import os
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def generate_reward(a):
    return np.random.normal(q[a], sigma_rew)


def pick_action():
    val = random.uniform(0, 1)
    if val > epsilon:
        action = np.argmax(Q)
    else:
        action = random.randint(0, num_actions - 1)
    return action


def pick_action_ucb(step):
    array = Q + c * np.sqrt(np.log(step)/num_picks)
    action = np.argmax(array)
    return action


def update_Q_samplemean(a):
    R = generate_reward(a)
    store = Q[a] * num_picks[a] + R
    num_picks[a] += 1
    Q[a] = store / num_picks[a]
    return R


def update_Q_constantalpha(a):
    R = generate_reward(a)
    Q[a] = Q[a] + const * (R - Q[a])
    return R


def update_Q_ucb(a):
    R = generate_reward(a)
    num_picks[a] += 1
    Q[a] = Q[a] + const * (R - Q[a])
    return R


def alter_q(q):
    q = q + np.random.normal(mu_q, sigma_q, num_actions)
    return q
```

```python
num_actions = 10
epsilon = 0.1
alpha = 0.1
num_steps = 10000
num_runs = 1000
sigma_rew = 1
mu_q = 0
sigma_q = 0.01
const = 0.1
x = np.arange(num_steps)
```

# Question 1

## Non Stationary, Average Reward

```python
globalrew_samplemean = np.zeros(num_steps, dtype=np.float64)

for run in range(num_runs):
    q = np.zeros(num_actions, dtype=np.float64)
    Q = np.zeros(num_actions, dtype=np.float64)
    num_picks = np.full(num_actions, 0)

    for step in range(num_steps):
        a = pick_action()
        R = update_Q_samplemean(a)
        globalrew_samplemean[step] += R
        q = alter_q(q)

globalrew_samplemean /= num_runs
```

```python
globalrew_constantalpha = np.zeros(num_steps, dtype=np.float64)

for run in range(num_runs):
    q = np.zeros(num_actions, dtype=np.float64)
    Q = np.zeros(num_actions, dtype=np.float64)

    for step in range(num_steps):
        a = pick_action()
        R = update_Q_constantalpha(a)
        globalrew_constantalpha[step] += R
        q = alter_q(q)

globalrew_constantalpha /= num_runs
```

```python
plt.plot(x[1:], globalrew_constantalpha[1:], label="Constant Alpha")
plt.plot(x[1:], globalrew_samplemean[1:], label="Sample Mean")
plt.xlabel("Steps")
plt.ylabel("Average Reward")
plt.legend()
plt.show()
```

## Non Stationary, % Optimal Action

```python
mat = np.zeros(shape=(num_runs, num_steps))

for run in range(num_runs):
    q = np.zeros(num_actions, dtype=np.float64)
    Q = np.zeros(num_actions, dtype=np.float64)
    num_picks = np.full(num_actions, 0)

    for step in range(num_steps):
        a = pick_action()
        if a == np.argmax(q):
            mat[run, step] = 1
        
        R = update_Q_samplemean(a)
        q = alter_q(q)

sums = np.sum(mat, axis = 0)
optimalaction_samplemean = sums/num_runs
```
